```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Dell\Downloads\archive\marketing_campaign.csv", sep='\t')
print("\nCustomer Personality Analysis Dataset:")
print("\nMissing Values:")
print(df.isnull().sum())
if 'Income' in df.columns:
    df['Income'] = df['Income'].fillna(df['Income'].median())

if 'Dt_Customer' in df.columns:
    df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y', errors='coerce')

df = df.dropna(subset=['Dt_Customer'])

for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.strip()

df = pd.get_dummies(df, columns=['Education', 'Marital_Status'], drop_first=True)

print("\nCleaned Data Preview:")
print(df.head())

df.to_csv(r"C:\Users\Dell\Downloads\archive\marketing_campaign_cleaned.csv", index=False)
```
